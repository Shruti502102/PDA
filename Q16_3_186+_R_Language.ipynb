```r
# Load required libraries
library(dplyr)
library(ggplot2)

# Sample dataset
data <- data.frame(
  year = c(2025, 2024, 2025, 2025, 2023, 2025, 2024),
  month = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul"),
  sales = c(1200, 900, 1500, 1300, 1000, 1800, 1100),
  category = c("Electronics", "Furniture", "Electronics", "Furniture", "Clothing", "Clothing", "Clothing")
)
```

```r
# Print original dataset
print("Original Dataset:")
print(data)
```

```r
# Step 1: Filter data for the year 2025
filtered_data <- data %>% filter(year == 2025)
print("Filtered Data (Year == 2025):")
print(filtered_data)
```

```r
# Step 2: Group by category and calculate average sales
avg_sales <- filtered_data %>%
  group_by(category) %>%
  summarise(average_sales = mean(sales))
print("Average Sales per Category:")
print(avg_sales)
```

```r
# Step 3: Plot bar chart
print("Displaying Bar Chart...")
ggplot(avg_sales, aes(x = category, y = average_sales, fill = category)) +
  geom_bar(stat = "identity") +
  labs(title = "Average Sales per Category (2025)",
       x = "Category",
       y = "Average Sales") +
  theme_minimal()
```
